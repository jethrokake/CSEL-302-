```python
import pandas as pd #Data manipulation
import numpy as np # numerical operation
import matplotlib.pyplot as plt #data visualization
from sklearn.linear_model import LinearRegression # DATA modeling
```

# 2. generating dataset

```python
data = pd.read_csv('Lesson_4_student_grades_dataset.csv')
data.head()
```

3. X and Y

```python
plt.scatter(data['Grades'], data['Attendance_Percentage'])
plt.xlabel('independent Variable (X)')
plt.ylabel('dependent Variable (Y)')
plt.title ('scatter plot of data')
plt.show()
```

```python
X = data[['Attendance_Percentage']] #independent
Y = data[['Grades']] #dependent
```

```python
model = LinearRegression()

model.fit(X,Y)
```

```python
slope = model.coef_[0]

inetercept= model.intercept_

print('', slope)
print('Intercept', inetercept)
```

```python
plt.scatter(X,Y)
plt.plot(X, model.predict(X), color = 'red')
plt.xlabel('dependent variable (X)')
plt.xlabel('independent variable (Y)')
plt.title('Measurment of Grades')
plt.show()
```

```python
new_X = [[70.09654790812836]]

predicted_Y = model.predict(new_X)

print('X= ', new_X)
print('predicted value of Y: ', predicted_Y)
```
